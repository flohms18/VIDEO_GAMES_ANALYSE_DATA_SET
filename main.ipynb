```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

df = pd.read_csv('vgsales.csv')

print(df.head())
```

```python
Platform_Sales_VG = df.groupby('Year')['Platform'].count()
Year_Sales = np.sort(df['Year'].drop_duplicates().dropna())


X360_Count = []
PS3_Count = []
Wii_Count = []

for index, row in df.iterrows():
    if row['Platform'] == 'Wii':
        Wii_Count.append(row['Name'])
    elif row['Platform'] == 'PS3':
        PS3_Count.append(row['Name'])
    elif row['Platform'] == 'X360':
        X360_Count.append(row['Name'])

ps3_games = df[df["Platform"] == "PS3"]
x360_games = df[df["Platform"] == "X360"]
wii_games = df[df["Platform"] == "Wii"]

wii_games = df[df["Platform"] == "Wii"]



best_selling_Wii = wii_games.sort_values(by="Global_Sales",ascending=False).head(1)
best_selling_x360 = x360_games.sort_values(by="Global_Sales",ascending=False).head(1)
best_selling_ps3 = ps3_games.sort_values(by="Global_Sales",ascending=False).head(1)
print(best_selling_x360['Name'])
print(best_selling_ps3['Name'])
print(best_selling_Wii['Name'])

print(wii_games['Global_Sales'].max())

Labels_Games = [best_selling_x360['Name'],best_selling_ps3['Name'],best_selling_Wii['Name']]

Count_Games = [wii_games['Global_Sales'].max(),ps3_games['Global_Sales'].max(),x360_games['Global_Sales'].max()])


plt.bar(Labels_Games,Count_Games)
plt.show()
Console = np.array(["Xbox 360","Wii","PS3"])
Game_count = np.array([len(X360_Count),len(Wii_Count),len(PS3_Count)])
    

print(len(X360_Count))
print(len(Wii_Count))
print(len(PS3_Count))






plt.bar(Console,Game_count,color="#23282d")
plt.title("Number of games sold on each platform with at least 100,000 copies")
plt.show()

plt.figure(facecolor='#F8F9FA')
ax = plt.gca()
ax.set_facecolor("#F8F9FA")

plt.plot(Year_Sales, Platform_Sales_VG, color="#23282d")
plt.show()


```

```python
best_selling_Wii = wii_games.sort_values(by="Global_Sales",ascending=False).head(1)
best_selling_x360 = x360_games.sort_values(by="Global_Sales",ascending=False).head(1)
best_selling_ps3 = ps3_games.sort_values(by="Global_Sales",ascending=False).head(1)

test_wii = best_selling_Wii.iloc[0]['Name']
test_x360 = best_selling_ps3.iloc[0]['Name']
test_ps3 = best_selling_x360.iloc[0]['Name']


bs_PS3 = best_selling_ps3.iloc[0]['Global_Sales']
bs_x360 = best_selling_x360.iloc[0]['Global_Sales']
bs_Wii = best_selling_Wii.iloc[0]['Global_Sales']

bs = [bs_Wii,bs_x360,bs_PS3]



print(bs_PS3)
print(test_wii)
print(test_x360)
print(test_ps3)

james = [test_wii,test_x360,test_ps3]
plt.figure(facecolor='#F8F9FA')
ax = plt.gca()
ax.set_facecolor("#F8F9FA")
plt.bar(james,bs,color="#23282d")

plt.title("Best-selling game for the Wii, PS3, Xbox 360 in millions")
plt.ylabel("Number of games sold in millions")

```

```python
df_PS3 = df[df["Platform"] == "PS3"]
df_X360 = df[df["Platform"] == "X360"]
df_Wii = df[df["Platform"] == "Wii"]

PS3_Global_sales = df_PS3["Global_Sales"].mean()
X360_Global_sales = df_X360["Global_Sales"].mean()
Wii_Global_sales = df_Wii["Global_Sales"].mean()
plt.figure(facecolor='#F8F9FA')
ax = plt.gca()
ax.set_facecolor("#F8F9FA")

print(PS3_Global_sales)
print(X360_Global_sales)
print(Wii_Global_sales)
m = ["PS3","X360","Wii"]
james = [PS3_Global_sales,X360_Global_sales, Wii_Global_sales]
plt.title("Average of games sold")
plt.ylabel("Unit sold in millions")

plt.bar(m,james,color="#23282d")
plt.show()
```
